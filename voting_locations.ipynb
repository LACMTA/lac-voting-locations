# Process Voting Locations for Third Party Consumption

- Vote By Mail Boxes
- Voting Centers

Vote By Mail Boxes are available from October 3, 2024 to 8:00pm November 5, 2024.


```python
import pandas as pd
import os

# Vote By Mail Boxes
VBM_PATH = os.path.normpath('lacrrcc-ballot-box-geodata.json')

# Voting Centers
#VC_PATH = os.path.normpath('lacrrcc-voting-centers.json')
```

## Vote By Mail Boxes

Setup dataframe.

```python
vbm_ogdf = pd.read_json(VBM_PATH)
vbm_mod = vbm_ogdf.copy()

vbm_mod.head()
```

### Remove Columns

Check whether there is more than one unique value for this set of columns.

```python
columns_to_drop = [
    'parkingArea',
    'isActive',
    'isCancelled',
    'votingPeriod',
    'waitTime',
    'waitTimeTS',
    'earlyVotingOpenTimeInPT',
    'earlyVotingCloseTimeInPT',
    'electionDayOpenTimeInPT',
    'electionDayCloseTimeInPT',
    'isInEarlyVotingRange',
    'isInElectionDayRange',
    'isWaitTimeUpdated',
    'isOpenButNotReported',
    'showLocationClosed',
    'electionDate',
    'earlyVotingStartDate',
    'earlyVotingEndtDate',
    'secondWaveEarlyVotingStartDate'
]

for column in columns_to_drop:
    if (vbm_ogdf[column].unique().size == 1):
        vbm_mod = vbm_mod.drop(columns=[column])

vbm_mod.head()
```

Remove the columns not needed by consumers, leaving only:

- `locationID`
- `name`
- `hours`
- `latitude`
- `longitude`

```python
# Remove extra columns

extra_columns = [
    'locationTypeCode',
    'electionID',
    'address',
    'city',
    'state',
    'zipCode',
    'additionalInformation'
]

vbm_mod = vbm_mod.drop(columns=extra_columns)

vbm_mod.head()
```

### Export Results

```python
# export to json
vbm_mod.to_json('la-county-rrcc-vbm.json', orient='records')
```
